```python
# conjunto set
# conjunto é uma coleção não ordenada de elementos únicos

# criando um conjunto
conjunto = {1, 2, 3, 3, 4, 5, 5, 5, 4, 6, 7, 7, 9}
print(conjunto)
```

```python
lista = [1, 1, 2, 3, 3, 4, 5, 5, 5, 4, 6, 7, 7]
valores_exclusivos = set(lista)
print(valores_exclusivos)
```

```python
# o método add() adiciona um elemento ao conjunto
conjunto.add(8)

# o método pop() remove um elemento aleatório do conjunto
conjunto.pop()

# método discard() remove um elemento específico do conjunto
conjunto.discard(5)
```

```python
print(conjunto)
```

```python
#  set também não ordena, precisamos do métood sorted() do python
ordenados = sorted(conjunto)
print(ordenados)

# vira uma lisata. Se quiser em sorted
ordenados = set(sorted(conjunto))
print(ordenados)


# for ordenado
for n in sorted(conjunto):
    print(n)
```

```python
lista_valores = []
for n in range(10):
    print(f'Informe o {n+1}º valor: ')
    num = int(input("Digite um número: "))
    lista_valores.append(num)
    

valores_exclusivos = set(lista_valores)
print(valores_exclusivos)
```
