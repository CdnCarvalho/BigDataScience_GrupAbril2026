```python
# dicionário com seus contatos
meus_contatos = {
    'Yan': '1234-5678',
    'Pedro': '9999-9999',
    'Ana': '8765-4321',
    'João': '8887-7778'
}

# dicionário com os contatos do Pedro
contatos_do_pedro = {
    'Yan': '1234-5678',
    'Fernando': '4345-5434',
    'Luiza': '4567-7654'
}

# percorre cada nome no dicionário contatos_do_pedro
for nome in contatos_do_pedro:
    
    # adiciona ou sobrescreve no dicionário meus_contatos
    meus_contatos[nome] = contatos_do_pedro[nome]

# exibe o resultado final
print(meus_contatos)
```
