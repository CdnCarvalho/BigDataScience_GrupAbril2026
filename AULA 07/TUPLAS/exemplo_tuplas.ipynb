```python
tupla_vazia = ()  # ou tuple()
tupla = (1, 2, 4, 2, 3, 1, 2, 5)
tupla_um_elemento = (8,)  # precisa da vírgula, se não, vira inteiro
```

```python
print(tupla_vazia)  # imprime tupla vazia
print(tupla)  # imprime todos os elementos da tupla
print(tupla[0])  # elemento da posição inicial 
print(tupla_um_elemento)
```

```python
nome = 'João'
idade = 45
altura = 1.68
pessoa = (nome, idade, altura)
print(pessoa)
print(type(pessoa))
```

```python
# métodos para tuplas
print(tupla.count(2))  # conta quantas vezes o elemento aparece na tupla
print(tupla.index(4))  # retorna a posição do elemento na tupla 
# tuplas são imutáveis, ou seja, não podemos alterar os elementos
# tupla[0] = 10  # isso vai dar erro
```

```python
# Tuplas não podem ser ordenadas. Se precisar, podemos usar o método sorted() do python e criar uma nova tupla
ordenados = sorted(tupla)  # Isto ordena, mas gera uma Lista ordenada
print(ordenados) 
print(type(ordenados))

# O ideal é que fosse feito assim, para continuar com os dados em tupla, pois o sorted retorna uma lista:
ordenados = tuple(sorted(ordenados))  # convertendo a lista para tupla
print(ordenados)
```

```python
# Ordenar em decrescente e converter para tupla
ordenados = tuple(sorted(tupla, reverse=True))
print(ordenados)
```

```python
# recebendo os dados de uma lista
produtos = ['celular', 'notebook', 'mouse', 'teclado']
# Convertendo a lista para tupla com o método tuple() do Python
tupla_produtos = tuple(produtos)
print(tupla_produtos)

lista_produtos = list(tupla_produtos)
print(lista_produtos)
```

```python
# verifica se um elemento está no conjunto
if 'pendrive' in produtos:
    print('Achei')



if 'pendrive' not in produtos:
    print('Não achei')
else:
    print('encontrei')
```

```python
# iterando sobre uma tupla
i = 1
for produto in tupla_produtos:
    print(f'produto {i}: {produto}')
    i += 1
```
