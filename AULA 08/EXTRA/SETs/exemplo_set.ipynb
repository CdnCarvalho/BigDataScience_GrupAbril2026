```python
# conjunto set
# conjunto é uma coleção não ordenada de elementos únicos, imutável em parte* e não indexada.
# imutável em parte* => Cada elemento não pode ser alterado.
# Não indexado, pois não possuem índice para acessar os elementos.
# Porém, podemos adicionar e remover elementos.

# criando um conjunto
conjunto = {12, 10, 1, 2, 3, 3, 4, 5, 5, 5, 4, 6, 7, 7, 9}
print(conjunto[0])  # erro: sets não possui índice
```

```python
lista = [20, 15, 1, 1, 2, 3, 3, 4, 5, 5, 5, 4, 6, 7, 7]
valores_exclusivos = set(lista)
print(valores_exclusivos)
```

```python
# Não é possível acessar itens em um conjunto pelo índice ou uma chave.

# o método add() adiciona um elemento ao conjunto
conjunto.add(8)

# o método pop() remove um elemento aleatório do conjunto
conjunto.pop()

# método discard() remove um elemento específico do conjunto
conjunto.discard(5)

print(conjunto)
```

```python
# Verifica se um elemento está no conjunto

# Não é possível acessar itens em um conjunto pelo índice ou uma chave.
# Mas você pode percorrer os itens do conjunto usando um for loop, ou verificar
# se um valor específico está presente em um conjunto, usando a inpalavra-chave.

print(5 in conjunto)

# se o elemento estiver no conjunto, multiplique por 2
if 5 in conjunto:
    print(f'5 Está presente e o dobro é: {5 * 2}')

# Verifica se um elemento não está no conjunto
if 13 not in conjunto:
    print('valor não encontrado')
```

```python
# Não é possível acessar itens em um conjunto pelo índice ou uma chave.
# Mas você pode percorrer os itens do conjunto usando um for loop, ou verificar
# se um valor específico está presente em um conjunto, usando a inpalavra-chave.

lista_valores = []
for n in range(10):
    num = int(input("Digite um número: "))
    lista_valores.append(num)
    

valores_exclusivos = set(lista_valores)
print(valores_exclusivos)
```
